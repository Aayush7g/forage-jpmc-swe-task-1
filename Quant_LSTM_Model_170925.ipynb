```python
"""
LSTM Stock Price Forecast - Google Colab Ready

This is a complete, well-documented Python "code-book" you can paste into Google Colab
(or save as a .py/.ipynb). It fetches historical stock prices using yfinance, builds
an LSTM model with Keras/TensorFlow, trains it and visualizes predictions.

Change the TICKER, date range, or hyperparameters at the top to experiment.

Steps included:
1. Setup & dependencies
2. Data download (yfinance)
3. Exploratory checks & visualization
4. Feature engineering (uses 'Close' but shows how to add features)
5. Train / validation split and sequence creation
6. LSTM model build, compile
7. Training with callbacks (EarlyStopping, ModelCheckpoint)
8. Evaluation (RMSE, MAE) and plots
9. Saving model & example of how to use it for future predictions

Run this in Google Colab. If you want a full .ipynb file exported, tell me and I can prepare it.

"""

# Install required packages (uncomment in Colab)
# !pip install yfinance --quiet
# !pip install tensorflow --quiet
# !pip install joblib --quiet

import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import joblib

# ---------------------------
# Config / Hyperparameters
# ---------------------------
TICKER = 'AAPL'                       # change to any ticker (e.g., 'MSFT', 'TSLA')
START_DATE = '2015-01-01'
END_DATE = datetime.today().strftime('%Y-%m-%d')
LOOKBACK = 60                         # how many past days the LSTM looks at
TEST_SIZE = 0.2                       # fraction of data used for test
BATCH_SIZE = 32
EPOCHS = 50
MODEL_DIR = 'saved_models'
SEED = 42

# reproducibility
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Create folder to save model
os.makedirs(MODEL_DIR, exist_ok=True)

# ---------------------------
# 1) Data Download
# ---------------------------
print(f"Downloading {TICKER} from {START_DATE} to {END_DATE}...")
raw = yf.download(TICKER, start=START_DATE, end=END_DATE, progress=False)
if raw.empty:
    raise SystemExit(f"No data found for {TICKER}. Try a different ticker or date range.")

# Keep only relevant features (Close, Open, High, Low, Volume)
df = raw[['Close', 'Open', 'High', 'Low', 'Volume']].copy()
df.dropna(inplace=True)
print(f"Downloaded {len(df)} rows.")

# Quick plot of Close price
plt.figure(figsize=(12,4))
plt.plot(df.index, df['Close'])
plt.title(f'{TICKER} Close Price')
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()

# ---------------------------
# 2) Feature scaling & sequences
# ---------------------------
scaler = MinMaxScaler(feature_range=(0,1))
scaled = scaler.fit_transform(df.values)

# helper to create sequences
def create_sequences(data, lookback=60):
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, :]) # Include all features in X
        y.append(data[i, 0]) # Predict only the Close price (index 0)
    X, y = np.array(X), np.array(y)
    # reshape X for LSTM -> (samples, timesteps, features)
    X = X.reshape((X.shape[0], X.shape[1], data.shape[1])) # Adjust shape to include all features
    return X, y

# train/test split index
split_idx = int(len(scaled) * (1 - TEST_SIZE))
train_scaled = scaled[:split_idx]
test_scaled = scaled[split_idx - LOOKBACK:]   # include lookback window

X_train, y_train = create_sequences(train_scaled, LOOKBACK)
X_test, y_test = create_sequences(test_scaled, LOOKBACK)

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)

# ---------------------------
# 3) Build the LSTM model
# ---------------------------
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(LOOKBACK, X_train.shape[2])), # Adjust input shape
    Dropout(0.2),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# ---------------------------
# 4) Callbacks and training
# ---------------------------
model_path = os.path.join(MODEL_DIR, f'{TICKER}_lstm_best.h5')
callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
    ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True, verbose=1)
]

# Recompile the model before training
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    shuffle=True,
    verbose=2
)

# ---------------------------
# 5) Evaluate & Predictions
# ---------------------------
# Predict on test set
preds_scaled = model.predict(X_test)

# invert scaling
# To inverse transform, we need to create a dummy array with the same number of features
dummy_array = np.zeros((len(preds_scaled), df.shape[1]))
dummy_array[:, 0] = preds_scaled.flatten() # Put the predictions in the Close price column
preds = scaler.inverse_transform(dummy_array)[:, 0] # Inverse transform and get the Close price

dummy_array_true = np.zeros((len(y_test), df.shape[1]))
dummy_array_true[:, 0] = y_test.flatten()
true = scaler.inverse_transform(dummy_array_true)[:, 0]


# Metrics
rmse = np.sqrt(mean_squared_error(true, preds))
mae = mean_absolute_error(true, preds)
print(f'Test RMSE: {rmse:.4f}')
print(f'Test MAE: {mae:.4f}')

# ---------------------------
# 6) Plot predictions vs true
# ---------------------------
# build a dataframe to align dates
test_dates = df.index[split_idx:]
results = pd.DataFrame({'Date': test_dates, 'True': true.flatten(), 'Predicted': preds.flatten()})
results.set_index('Date', inplace=True)

plt.figure(figsize=(14,5))
plt.plot(results['True'], label='True')
plt.plot(results['Predicted'], label='Predicted')
plt.title(f'{TICKER} - True vs Predicted Close Price')
plt.legend()
plt.show()

# Zoom last 120 days
plt.figure(figsize=(14,5))
plt.plot(results['True'][-120:], label='True')
plt.plot(results['Predicted'][-120:], label='Predicted')
plt.title(f'{TICKER} - Zoom last 120 days')
plt.legend()
plt.show()

# ---------------------------
# 7) Save model and scaler for future use
# ---------------------------
scaler_path = os.path.join(MODEL_DIR, f'{TICKER}_scaler.joblib')
joblib.dump(scaler, scaler_path)

# Save model in Keras format
model.save(os.path.join(MODEL_DIR, f'{TICKER}_lstm_full.keras'))
print('Saved model and scaler to', MODEL_DIR)

# ---------------------------
# 8) Example: single-step future prediction (next day)
# ---------------------------
# take the last LOOKBACK days from the full scaled dataset
last_window = scaled[-LOOKBACK:]
last_window = last_window.reshape((1, LOOKBACK, scaled.shape[1])) # Adjust shape
next_pred_scaled = model.predict(last_window)

# Inverse transform the prediction
dummy_array_pred = np.zeros((1, df.shape[1]))
dummy_array_pred[:, 0] = next_pred_scaled.flatten()
next_pred = scaler.inverse_transform(dummy_array_pred)[:, 0]

print(f'Next day predicted Close price for {TICKER}: {next_pred.flatten()[0]:.4f}')

# ---------------------------
# Notes & Next steps
# ---------------------------
# - This notebook uses only Close price. You can add features: volume, technical indicators
#   (SMA, RSI, MACD), macro variables, or alternative data (news sentiment).
# - Consider using walk-forward validation (time-series CV) rather than a single train/test split.
# - Try hyperparameter tuning (units, layers, learning rate, lookback) using Keras Tuner or Optuna.
# - Beware of lookahead bias and data leakage. Always construct sequences using only past data.
# - This LSTM predicts prices directly. Some workflows predict returns or price ratios instead.

# End of file
```
